# What is log_loss? and How to calculate it?


![Getting Started](./logloss.png)

Evalution metric used for loss calculation using true label and predicted probability of data_point belonging to perticular class.
When we use the logreg.predict_proba(x_data_point), this calculates the probability of x_data_point belonging to perticular class. For example X_daata_point has true_label 1 and the output of logreg.predict_proba(x_data_point) is 0.8. This means 80% the x_data_point belong to class 1. Then y become 1 and p become 0.8 in above formula. this is how we do calculate the prediction probability.

The Output could be interpreted as following.

![Getting Started](./results.PNG)

```python
TotalPredictPro=pd.DataFrame(data=[],index=trainy.index,columns=[0,1])
Score_train=[]
Score_cv=[]
for train_index,cv_index in k_fold.split(trainx,trainy):
    train_x_fold, x_cv_fold = trainx.iloc[train_index,:], trainx.iloc[cv_index,:]
    train_y_fold, y_cv_fold = trainy.iloc[train_index], trainy.iloc[cv_index]

    # fit the model
    logReg.fit(train_x_fold,train_y_fold)

    predictpro_tr=logReg.predict_proba(train_x_fold)  # predict_proba is confidence score of detected value be;ongs to perticular class
    
    # calculate the loss
    entropy_loss_tr=log_loss(train_y_fold,predictpro_tr)
    
    Score_train.append(entropy_loss_tr)
    

    # performance evaluation on the validation dataset
    predictpro_val=logReg.predict_proba(x_cv_fold)
    entropy_loss_val=log_loss(y_cv_fold,predictpro_val)
    Score_cv.append(entropy_loss_val)


    TotalPredictPro.loc[x_cv_fold.index,:] =logReg.predict_proba(x_cv_fold)


    print("******************************************************")
    print(f"The log_loss on training dataset is { entropy_loss_tr}")
    print(f"The log_loss on validation dataset is { entropy_loss_val}")

print("**********************************************")
predicty=logReg.predict_proba(trainx)
total_entropy_loss=log_loss(trainy,predicty)

print(f"Total log_loss given as {total_entropy_loss}")
```

```python
# Number of trees in the forest:
n_estimators = 15
# Number of features to consider when looking for the best split:
max_features="sqrt"
# Maximum depth of the tree:
max_depth = 20
# Minimum number of samples required to split an internal node:
min_samples_split = 3
# Minimum number of samples required to be at a leaf node:
min_samples_leaf = 1
# Grow trees with max_leaf_nodes in best-first fashion. Best nodes are defined as relative reduction in impurity. 
max_leaf_nodes = None
# Whether bootstrap samples are used when building trees. If False, the whole dataset is used to build each tree:
bootstrap = False
# Whether to use out-of-bag samples to estimate the generalization score. Only available if bootstrap=True.
oob_score = False
# Number of jobs to run in parallel. (-1) means use all.
n_jobs = -1
# Random state
random_state = 2020
#-----------------------------------------------------------------
# Creating the classifier:
RFC = RandomForestClassifier(n_estimators=n_estimators, max_features=max_features, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, \
                              max_leaf_nodes=max_leaf_nodes, bootstrap=bootstrap,oob_score=oob_score, n_jobs=n_jobs, random_state=random_state)
```

```python
model1=RFC

trainScore=[]
valScore=[]
total_prob=pd.DataFrame(data=[],index=trainx.index,columns=[0,1])


for train_index,cv_index in k_fold.split(trainx,trainy):
    train_x_fold, x_cv_fold = trainx.iloc[train_index,:], trainx.iloc[cv_index,:]
    train_y_fold, y_cv_fold = trainy.iloc[train_index], trainy.iloc[cv_index]

    model1.fit(train_x_fold,train_y_fold)
    # let look at prediction probability of being 0(neg) or 1(pos) for training fold
    prob_train=model1.predict_proba(train_x_fold)
    loglosstrain=log_loss(train_y_fold,prob_train)
    trainScore.append(loglosstrain)

    # let look at prediction probability of being 0(neg) or 1(pos) for cv fold
    prob_cv=model1.predict_proba(x_cv_fold)
    loglosscv=log_loss(y_cv_fold,prob_cv)
    valScore.append(loglosscv)


    total_prob.loc[x_cv_fold.index,:]=model1.predict_proba(x_cv_fold)
    
    print("**************************************")
    print(f"Training Log Loss is {loglosstrain}")
    print(f"CV Log Loss is {loglosscv}")

pre_train_y=model1.predict_proba(trainx)
loglosstotal=log_loss(trainy,pre_train_y) 

print("************************************")
print(f"RFC log loss {loglosstotal}")
```

So as we are considering the logistic regression, mean the signois activation function applied on the top of linear regression. By default we have considered the threshold as 0.5 mean the results or confusion matrix has been shown for the threshold of 0.5.
We can change hte threshold simply by __predict >0.7__. So for different threshold we will have different precision and recall. That is what we are plotting using precision-recall curve.

__How to read/understabt the precision-recall curve?__

As we see initially the precision is high 1 (100% precision), as we classifiy more and more data points the precision start changing. When the threshold change the precision and recall changes. for threshold 1 we have recall 100% mean there is no __False Negative__ then we can say this is really good we don't have any FN. But what about the precision? do we have good precisiona for same threshold? __NO, RIGHT__ 

Then what would be our approach? we want to have a such threshold values for whcihc we can have good precision aand recall (Again this is also depends on the use cases and probelm statement.)

In case of driver safety we could ahve chosen the threshold between 0.8 to 0.9. For this values of threshold we have good precision and Recall. In other case such as the Maintannce cost reduction we could have chosen the threshold value for which we could have high precision such as 0.75 to 0.9.

__Receiver operating characteristic curve?__

TPR=TP/TP+FN  called recalled and in other words __Sensivity__ which is defined as the ability of defined model to detect the positives given the true positives. __TPR=(1-FPR)__

FPR=FP/FP+TN

AUC the ROC curve measures the a classifier's/model's ability to classify the different classes for the various threshold values. 

ROC is performance metric, to evaluate the performace of the model ROC is used. When area under curve is 1 means the model performing at its best.


## Machine learning Approch: adaboost

__What is adaboost?__

It is the type of Boosting algorithm. Boosting is the ensemble method, covert weak learner to strong one. The weak learner are aranged in the series. 

The weak learners are the decision trees with one level called as stump
